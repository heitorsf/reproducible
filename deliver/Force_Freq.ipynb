Abaixo são plotadas as curvas.

```python
Freqs = []
Forces = []
fig = plt.figure()
plt.xlabel(u"Frequência de disparo [Hz]")
plt.ylabel(u"Força tetânica [N]")
for i,mutype in enumerate(['S','FR','FF']):
    Freqs.append(np.loadtxt("freq"+mutype+".txt"))
    Forces.append(np.loadtxt("force"+mutype+".txt"))
    plt.plot(Freqs[i],Forces[i],label="Tipo "+mutype)
plt.legend()
plt.show()
import os
basewd = os.getcwd()
os.chdir('/work/reproducible/figures/notebook')
fig.savefig("Force_Freq.png")
os.chdir(basewd)
```
