### Descriçãos dos dados

Neste estudo foram implementados modelos bicompartimentais de neurônios motores medulares. Foi utilizado um conjunto de dados que representam os parâmetros de cada neurônio.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
basedir = os.getcwd()
os.chdir('..')
os.chdir('data')
data = pd.read_csv("Spinal_Cord_Motor_Neurons_2019-04-16_11-09-06.csv")
os.chdir(basedir)
```

```python
print(data)
```

```python
print(data.keys())
```

```python
Soma_gnabar_napp = data.Soma_gnabar_napp.values
```

```python
numNeuronios = len(Soma_gnabar_napp)
```

```python
plt.figure()
plt.title("Condutância específica de íons Sódio")
plt.xlabel("# neurônio")
plt.ylabel("gNa [mS/cm²]")
plt.plot(np.arange(numNeuronios),Soma_gnabar_napp,'.')
plt.show()
```
